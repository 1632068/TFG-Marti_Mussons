```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("parelles_mal_classificades.csv")
```

```python
sns.histplot(df["similitud"], bins=30, kde=True)
plt.axvline(0.5, color='red', linestyle='--', label="Llindar 0.5")
plt.title("Distribució de les similituds en parelles mal classificades")
plt.xlabel("Similitud cosinus")
plt.ylabel("Freqüència")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df["error_tipus"] = np.where((df["label_real"] == 0) & (df["prediccio"] == 1), "Fals positiu",
                             np.where((df["label_real"] == 1) & (df["prediccio"] == 0), "Fals negatiu", "Altres"))

sns.histplot(data=df, x="similitud", hue="error_tipus", kde=True, bins=30)
plt.title("Similitud segons el tipus d'error")
plt.xlabel("Similitud cosinus")
plt.grid(True)
plt.show()
```

```python
# Separem falsos positius i negatius
fp = df[(df["label_real"] == 0.0) & (df["prediccio"] == 1)]
fn = df[(df["label_real"] == 1.0) & (df["prediccio"] == 0)]

# Mostrem una mostra aleatòria de cada grup
print("\n Falsos Positius (classificats com similars, però no ho són):")
print(fp.sample(3, random_state=42)[["text_a", "text_b", "label_real", "prediccio", "similitud"]].to_string(index=False))

print("\n Falsos Negatius (classificats com diferents, però eren similars):")
print(fn.sample(3, random_state=42)[["text_a", "text_b", "label_real", "prediccio", "similitud"]].to_string(index=False))
```
